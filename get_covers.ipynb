```js
const collectedSrcset = [];

document.querySelectorAll('picture').forEach(picture => {
    const source = picture.querySelector('source');
    if (source) {
        const srcset = source.srcset.split(",").pop().trim();
        collectedSrcset.push(srcset);
        source.remove();
    }
});

if (collectedSrcset.length > 0) {
    const textContent = collectedSrcset.join("\n");
    const timestamp = new Date().toISOString().replace(/[:.]/g, "-");
    const filename = `srcset-${timestamp}.txt`;

    const blob = new Blob([textContent], { type: "text/plain" });
    const link = document.createElement("a");
    link.href = URL.createObjectURL(blob);
    link.download = filename;
    document.body.appendChild(link);
    link.click();
    document.body.removeChild(link);
}
```

在console手动加载图片后多运行几次，python中有去重。放入`./album_list`

````python
import os
````

````python
album_list = []
for i in os.listdir('./album_list'):
    with open('./album_list/' + i, 'r') as f:
        album_list.extend(f.readlines())
album_list = [i.split(' ')[0] for i in album_list]
album_list = list(set(album_list))
album_list
````

````python
import os
import requests

# 创建一个目录来存储下载的图片
os.makedirs('./album_cover', exist_ok=True)

# 假设 album_list 中是图片的 URL
for i, url in enumerate(album_list):
    try:
        # 发送请求获取图片
        response = requests.get(url)
        response.raise_for_status()  # 检查请求是否成功

        # 获取图片的文件名
        file_name = os.path.join('./album_cover', f"image_{i+1}.jpg")

        # 保存图片到本地
        with open(file_name, 'wb') as f:
            f.write(response.content)

        print(f"Downloaded image {i+1} from {url}")
    except requests.exceptions.RequestException as e:
        print(f"Failed to download {url}: {e}")
````
